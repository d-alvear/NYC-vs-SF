```python
import pandas as pd
import numpy as np
import json
import folium
from folium import plugins
```

```python
with open('NYC_Neighborhood_Tabulation_Areas.geojson') as f:
    nyc_data = json.load(f)
```

```python
nyc_data.keys()
```

```python
nyc_data["features"][0]
```

```python
import itertools

features = []
for k,v in itertools.groupby([i for i in nyc_data['features'] if not i['properties']['boro_name'] != 'Manhattan' ]):  # filter out city2
    features.append(k)
nyc_data['features'] = features
```

```python
nyc_data
```

```python
m = folium.Map(
    location=[40.7128, -74.0060],
    zoom_start=12
)

folium.GeoJson(
    nyc_data,
    name='geojson'
).add_to(m)

m
```

```python
with open('nyc_neighborhoods.json', 'w') as f:
    json.dump(nyc_data, f)
```
