## Mapping Notebook

```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
import json
```

### Rendering San Francisco Map

```python
with open('SF_Find_Neighborhoods.geojson') as f:
    sf_data = json.load(f)
```

```python
m = folium.Map(
    location=[37.7749, -122.4194],
    zoom_start=12
)

folium.GeoJson(
    sf_data,
    name='geojson'
).add_to(m)

m
```
