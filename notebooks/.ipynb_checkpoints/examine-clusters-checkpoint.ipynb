```python
cluster_eight = sf_clusters.loc[sf_clusters['Cluster Label'] == 8, sf_clusters.columns[[0] + list(range(4, sf_clusters.shape[1]))]]
cluster_eight.iloc[0].unique()
```

```python
categories_freq = {}
temp = cluster_eight.drop('Neighborhood', axis=1)
temp.iloc[0].unique()

for i in range(len(temp)):
    venues = temp.iloc[i].unique()
    
    for v in venues:
        if v not in categories_freq:
            categories_freq[v] = 1
        if v in categories_freq:
            categories_freq[v] += 1
```

```python
sorted(categories_freq, key=categories_freq.get, reverse=True)[:5]
```
