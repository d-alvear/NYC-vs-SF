```python
import pandas as pd
import numpy as np
import json
import folium
from folium import plugins
```

```python
with open('nyc_neighborhoods.json') as f:
    nyc_data = json.load(f)
```

```python
with open('ZillowNeighborhoods-NY.geojson') as f:
    zillow_data = json.load(f)
```

```python
import itertools

features = []
for k,v in itertools.groupby([i for i in zillow_data['features'] if not i['properties']['County'] != 'New York' ]):  # filter out city2
    features.append(k)
zillow_data['features'] = features
```

```python
zillow_data["features"][10]["properties"]
```

```python
nyc_data
```

```python
m = folium.Map(
    location=[40.7128, -74.0060],
    zoom_start=12
)

folium.GeoJson(
    nyc_data,
    name='orig_geojson'
).add_to(m)


folium.LayerControl().add_to(m)

m
```

```python
m = folium.Map(
    location=[40.7128, -74.0060],
    zoom_start=20
)

folium.GeoJson(
    zillow_data,
    name='zillow_geojson'
).add_to(m)

m
```

```python
with open('zillow-nyc.geojson', 'w') as f:
    json.dump(zillow_data, f)
```
