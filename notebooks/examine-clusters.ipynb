```python
import pandas as pd 
import numpy as np

import matplotlib.cm as cm
import matplotlib.colors as colors
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sf_clusters = pd.read_csv('../data/exported-data/sf_clusters.csv',index_col=0)
nyc_clusters = pd.read_csv('../data/exported-data/nyc_clusters.csv',index_col=0)

print(sf_clusters.columns, nyc_clusters.columns)
```

Here I define a function that will tell me the common venues based on the cluster.

```python
def define_cluster(cluster_num):

    cluster = sf_clusters.loc[sf_clusters['Cluster Label'] == cluster_num, sf_clusters.columns[[0] + list(range(4, sf_clusters.shape[1]))]]

    categories_freq = {}
    temp = cluster.drop('Neighborhood', axis=1)
    temp.iloc[0].unique()

    for i in range(len(temp)):
        venues = temp.iloc[i].unique()
    
        for v in venues:
            if v not in categories_freq:
                categories_freq[v] = 1
            if v in categories_freq:
                categories_freq[v] += 1
            
    return (sorted(categories_freq, key=categories_freq.get, reverse=True)[:5])
```
